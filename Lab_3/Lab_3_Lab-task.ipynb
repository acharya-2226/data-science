# Data Visualization Practice Questions

## 1. Line Chart Practice Question
**Question:**
You are given a dataset of monthly rainfall (in mm) for your city over the past year. Create a line chart that shows the rainfall trend for the year. Label the axes and provide a title.

**Dataset**
- January: 45mm
- February: 38mm
- March: 60mm
- April: 70mm
- May: 75mm
- June: 80mm
- July: 90mm
- August: 85mm
- September: 72mm
- October: 50mm
- November: 40mm
- December: 30mm

**Instructions:**
- Label the x-axis as "Month" and the y-axis as "Rainfall (mm)".
- Title the chart as "Monthly Rainfall Trend Created by [Your Name]".

---



````python
# Import the necessary libraries
import matplotlib.pyplot as plt  # For creating visualizations

# Define the sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']  # List of months
rain = [45, 38, 60, 70, 75, 80, 90, 85, 72, 50, 40, 30]  # Corresponding sales data for each month

# Create a new figure for the plot with a specified size
plt.figure(figsize=(16, 6))  # The plot will be 10 inches wide and 6 inches tall

# Plot the data on a line chart
plt.plot(
    months,             # x-axis data (months)
    rain,              # y-axis data (rainfall)
    marker='o',         # Adds circular markers at each data point
    color='b',            # Line color (blue)
    linestyle='-',       # Line style (solid line)
    label='R'            # Label for the legend
)

# Add a title to the plot
plt.title('Monthly Rainfall Trend Chart By Siddharth Acharya')

# Label the x-axis
plt.xlabel('Months')

# Label the y-axis
plt.ylabel('Rainfall (in mm)')

# Add a legend to the plot
plt.legend()

# Add a grid to the plot for better readability
plt.grid(True)

# Display the plot
plt.show()
````

# Monthly Rainfall Trend Chart By Siddharth Acharya

This chart visualizes the monthly rainfall data over a year. The x-axis represents the months, and the y-axis represents rainfall (in mm). Circular markers are added for data points, and the chart includes a legend, grid, and proper labels for clarity.

## 2. Bar Chart Practice Question
**Question:**
You are given data showing the number of books read by students in your class during the past month. Create a bar chart to compare the number of books read by each student.

**Dataset**
- Student A: 5 books
- Student B: 7 books
- [Your Name]: 3 books
- Student D: 8 books
- Student E: 6 books

**Instructions:**
- Label the x-axis with the student names and the y-axis with the number of books.

---



````python
# Create a bar chart for book history of 5 students:

import matplotlib.pyplot as plt

# Products data
students = ['Student A', 'Student B', 'Siddharth', 'Student D', 'Student E']
book = [5, 7, 3, 8, 6]

# Creating the bar chart
plt.bar(students, book, color='skyblue')
plt.title("Books Read by Students")
plt.xlabel('←---------STUDENTS--------→')
plt.ylabel('←-------BOOKS--------→')
plt.show()
````

## 3. Heatmap Practice Question
**Question:**
You are given data showing the relationship between study hours and grade scores for 5 students. Create a heatmap to show how study time and grades are correlated.

**Dataset**
| Student   | Study Hours | Grade Score |
|-----------|-------------|-------------|
| Student 1 | 5           | 75          |
| Student 2 | 3           | 65          |
| Student 3 | 8           | 85          |
| [Your Name] | 2           | 60          |
| Student 5 | 6           | 80          |

**Instructions:**
- Create a heatmap with "Study Hours" on the x-axis and "Grade Score" on the y-axis.
- Label the axes and add a color scale.

---


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the dataset
data = {
    'Student': ['Student 1', 'Student 2', 'Student 3', 'Siddharth', 'Student 5'],
    'Study Hours': [5, 3, 8, 2, 6],
    'Grade Score': [75, 65, 85, 60, 80]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the correlation matrix
corr = df[['Study Hours', 'Grade Score']].corr()

# Plot the heatmap
plt.figure(figsize=(6, 4))  # Set the figure size
sns.heatmap(
    corr,
    annot=True,          # Annotate the cells with the correlation values
    cmap='plasma',     # Use the 'coolwarm' colormap for a visual gradient
    fmt='.3f',           # Format the numbers to 2 decimal places
    cbar=True            # Include a color bar
)
plt.title('Correlation Heatmap: Study Hours vs Grade Score')  # Add a title
plt.show()  # Display the heatmap
````

# Correlation Heatmap: Study Hours vs Grade Score

This heatmap visualizes the correlation between two variables in the dataset: *Study Hours* and *Grade Score*. The heatmap provides insight into the strength and direction of the relationship between these variables.

## Key Highlights:
1. **Data Overview**:
   - The dataset includes five students, including "Siddharth".
   - Variables analyzed are:
     - *Study Hours*: The number of hours each student studied.
     - *Grade Score*: The respective grade scores achieved.

2. **Correlation Matrix**:
   - The correlation coefficient is calculated using Pandas' `.corr()` method.
   - Values range from `-1` to `1`:
     - A value close to `1` indicates a strong positive correlation.
     - A value close to `-1` indicates a strong negative correlation.
     - A value near `0` indicates no significant relationship.

3. **Heatmap Visualization**:
   - **Annotations**: Correlation values are displayed within the cells (`annot=True`).
   - **Colormap**: The `'plasma'` gradient provides a clear visual representation of the correlation strength.
   - **Title**: `'Correlation Heatmap: Study Hours vs Grade Score'` emphasizes the focus.

This visualization helps identify how strongly *Study Hours* relate to *Grade Score*. For example, a high positive correlation would suggest that more study hours typically result in better grades.


## 4. Box Plot Practice Question
**Question:**
You have the exam scores of 10 students in your class. Create a box plot to show the distribution of exam scores.

**Dataset**
- 50, 65, 70, 85, 60, 75, 80, 90, 55, 70.

**Instructions:**
- Create a box plot to show the spread of the exam scores.
- Label the x-axis as "Exam Scores" and provide a title such as "Distribution of Exam Scores Created by [Your Name]".



````python

import seaborn as sns
import numpy as np

# Sample data
exam_scores = [50, 65, 70, 85, 60, 75, 80, 90, 55, 70]

# Plot boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(exam_scores, color='aqua')
plt.title('Distribution of Exam Scores Created by Siddharth Acharya')
plt.xlabel('←---------EXAM_SCORES--------→')

plt.show()
````

# Boxplot: Distribution of Exam Scores

This boxplot visualizes the distribution of exam scores for a group of students. Boxplots are useful for identifying the spread of data, including the median, quartiles, and potential outliers.

## Key Features:
1. **Data**: 
   - The dataset represents a sample of exam scores, including values like 50, 65, 70, 85, 60, etc.
   
2. **Boxplot Components**:
   - **Box**: The box in the plot represents the interquartile range (IQR), showing where the middle 50% of the data falls.
   - **Median**: The line inside the box shows the median score, which divides the dataset into two equal halves.
   - **Whiskers**: The lines extending from the box show the range of scores within a certain threshold (typically 1.5 times the IQR).
   - **Outliers**: Any points outside the whiskers are considered potential outliers.

3. **Plot Customizations**:
   - **Color**: The box is colored in `aqua` for an appealing visual effect.
   - **Title**: `'Distribution of Exam Scores Created by Siddharth Acharya'` gives context to the plot.
   - **X-Axis Label**: The label `'←---------EXAM_SCORES--------→'` is customized for clarity.

This boxplot provides a clear view of the distribution, including the spread, central tendency (median), and outliers in the exam scores dataset.



## 5. Scatter Plot Practice Question
**Question:**
You are given data showing the heights (in cm) and weights (in kg) of your friends. Create a scatter plot to show how weight and height are related.

**Dataset**
| Friend   | Height (cm) | Weight (kg) |
|----------|-------------|-------------|
| Friend 1 | 160         | 55          |
| Friend 2 | 170         | 62          |
| Friend 3 | 175         | 70          |
| Friend 4 | 165         | 58          |
| [Your Name] | 180         | 80          |

**Instructions:**
- Plot height on the x-axis and weight on the y-axis.
- Label the axes and title the chart as "Height vs Weight".
```

````python
import pandas as pd
import matplotlib.pyplot as plt

# Define the dataset
data = {
    'Friend': ['Friend 1', 'Friend 2', 'Friend 3', 'Friend 4', 'Siddharth'],
    'Height (cm)': [160, 170, 175, 165, 180],
    'Weight (kg)': [55, 62, 70, 58, 80]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(df['Height (cm)'], df['Weight (kg)'], color='blue', edgecolor='black', s=100)

# Add labels and title
plt.xlabel('Height (cm)')  # X-axis label
plt.ylabel('Weight (kg)')  # Y-axis label
plt.title('Height vs Weight')  # Title

# Annotate each point with the friend's name
for i, friend in enumerate(df['Friend']):
    plt.text(df['Height (cm)'][i] + 1, df['Weight (kg)'][i], friend, fontsize=10)

# Add a grid
plt.grid(True)

# Show the plot
plt.show()
````

# Scatter Plot: Height vs Weight of Friends

This scatter plot visualizes the relationship between the height and weight of five individuals, including "Siddharth." Each point represents a friend, and the position of each point on the graph corresponds to their height and weight.

## Key Features:
1. **Data Overview**:
   - The dataset includes five friends, with their height in centimeters and weight in kilograms.
   - Friend names are used as labels for each data point.

2. **Scatter Plot Components**:
   - **Points**: Each friend's height and weight are represented as a blue dot with black edges (`color='blue', edgecolor='black'`).
   - **Size**: The size of each point is increased to make the points more visible (`s=100`).
   - **Labels**: Each point is annotated with the corresponding friend's name for easy identification.

3. **Customization**:
   - **Grid**: A grid is added to make the chart more readable (`plt.grid(True)`).
   - **Labels and Title**: The axes are labeled with `'Height (cm)'` and `'Weight (kg)'`, and the title `'Height vs Weight'` provides context to the plot.

4. **Annotations**:
   - Each data point is labeled with the friend's name, positioned slightly to the right of each point, making the chart informative and easy to read.

This scatter plot provides insights into the relationship between height and weight for the individuals in the dataset. It can also be used to detect any patterns or outliers, such as Siddharth's position relative to others in terms of height and weight.
